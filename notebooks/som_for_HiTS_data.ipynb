```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import sompy

%matplotlib inline
sb.set(style="white", color_codes=True, context="notebook", font_scale=1.4)

mainpath = '/Users/jorgetil/Astro/HITS'
```

```python
labeled_set = pd.read_csv('%s/tables/HiTS_training_set_full_DA_goodP.csv' % (mainpath))
```

```python
cols = labeled_set.columns.values
FATS_feats_all = cols[10:]
FATS_feats_non = cols[10:-3]
print cols

print FATS_feats_non
```

```python
data_label = labeled_set[FATS_feats_non].values
print data_label.shape
```

```python
output_shape = (40, 40)
som = sompy.SOM(output_shape, data_label)
som.set_parameter(neighbor=0.1, learning_rate=0.2)

output_map = som.train(10000)
```

```python
print output_map[:,:,0].shape
```

```python
for k in range(output_map.shape[2]):
    plt.figure(figsize=(4,4))
    plt.title(FATS_feats_non[k])
    plt.imshow(output_map[:,:,k], interpolation='none', cmap='winter')
    plt.show()
```
